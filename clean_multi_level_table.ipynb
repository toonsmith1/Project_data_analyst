#**read excel file**



```python
#read excel file
df=pd.read_csv(path_file)
df
```

#**Test drop empty rows**


```python
#test drop blank row and reset index
df.dropna(subset=['Product Name']).reset_index(drop=True)
```

```python
df.head(50)
```

```python
#drop blank row and reset index
df=df.dropna(subset=['Product Name']).reset_index(drop=True)
df
```

```python
df=df.drop_duplicates().reset_index(drop=True)
df
```

```python
#check info
df.info()
```

```python
df.loc[0,:]
```

```python
excel_file_path = 'cleaned_data.xlsx'
df.to_excel(excel_file_path,index=False)
```

```python
#create new column Category
#df['Category']=df.apply(lambda r: r['Product Name'] if np.isnan(r['Calories'])else np.nan, axis=1)
df['Category'] = np.where(pd.isna(df['Calories']), df['Product Name'], np.nan)
df
```

```python
df['Category']=df['Category'].ffill()
df
```

```python
df=df.dropna(subset=['Calories']).reset_index(drop=True)
df
```
